## Carregando o dataset do disco local

O arquivo `impacto_ia_estudantes.csv` está na pasta `C:\projetohit`.

```python
import pandas as pd

# Usando caminho absoluto
df = pd.read_csv(r"C:\projetohit\impacto_ia_estudantes.csv")

# Mostrar as primeiras linhas
df.head()
```

## Explorando informações gerais do dataset

Nesta etapa vamos verificar:
- Quantidade de linhas e colunas
- Tipos de dados de cada coluna
- Valores nulos
- Estatísticas descritivas

```python
# Quantidade de linhas e colunas
print("Formato do dataset:", df.shape)

# Tipos de dados
print("\nTipos de dados:")
print(df.dtypes)

# Valores nulos
print("\nValores nulos por coluna:")
print(df.isnull().sum())

# Estatísticas descritivas
print("\nEstatísticas descritivas:")
print(df.describe(include="all"))
```

## Primeira pista da caça ao tesouro

Pergunta: Qual ferramenta de IA é a mais usada pelos estudantes?

Objetivo: Descobrir a campeã entre as ferramentas de IA.

```python
# Contagem das ferramentas de IA usadas
ferramentas = df["AI_Tool_Used"].value_counts()

# Mostrar o resultado
print("Ferramentas mais usadas:")
print(ferramentas)

# Mostrar só a campeã
print("\nA ferramenta mais usada é:", ferramentas.idxmax())
```

## Segunda pista da caça ao tesouro

Pergunta: O propósito de uso da IA (estudo vs. lazer) influencia o impacto nas notas?

Objetivo: Descobrir se o motivo pelo qual os estudantes usam IA muda os resultados acadêmicos.

```python
# Agrupar por propósito de uso e impacto nas notas
impacto_por_proposito = df.groupby("Purpose")["Impact_on_Grades"].value_counts()

print("Impacto nas notas por propósito de uso:")
print(impacto_por_proposito)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dados das cidades e quantidade de estudantes
dados_cidades = {
    "City": ["Delhi", "Chennai", "Bangalore", "Jaipur", "Pune", "Kolkata",
             "Lucknow", "Multan", "Ahmedabad", "Hyderabad", "Mumbai",
             "Lahore", "Islamabad", "Faisalabad", "Karachi"],
    "Quantidade": [37, 32, 32, 32, 30, 30,
                   29, 29, 28, 27, 23,
                   22, 18, 17, 14]
}

# Criar DataFrame
df_cidades = pd.DataFrame(dados_cidades)

# Ordenar do maior para o menor
df_cidades = df_cidades.sort_values(by="Quantidade", ascending=True)

# Criar gráfico de barras horizontais
plt.figure(figsize=(10,6))
plt.barh(df_cidades["City"], df_cidades["Quantidade"], color="skyblue")
plt.title("Quantidade de estudantes que usam IA por cidade")
plt.xlabel("Quantidade de estudantes")
plt.ylabel("Cidade")
plt.show()
```

## Quarta pista da caça ao tesouro

Pergunta: Qual faixa etária usa mais IA?

Objetivo: Identificar a idade ou grupo etário com maior número de estudantes que utilizam IA.

## Sexta pista da caça ao tesouro

Pergunta: Qual curso usa mais IA?

```python
import matplotlib.pyplot as plt

# Dados já calculados
cursos = df["Education_Level"].value_counts()

# Gráfico de barras verticais
plt.figure(figsize=(8,6))
cursos.plot(kind="bar", color="green")

# Títulos e rótulos
plt.title("Estudantes que usam IA por nível de educação")
plt.ylabel("Quantidade de estudantes")
plt.xlabel("Nível de educação")

# Mostrar valores acima das barras
for i, valor in enumerate(cursos):
    plt.text(i, valor + 2, str(valor), ha="center")

plt.show()
```
